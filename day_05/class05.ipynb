```python
inp = int(input("Enter a number: "))
if inp>12:
    print("Class Over")
else: 
    print("Wait for class to finish!")
```

```python
try:
    inp = int(input("Enter a number: "))
    if inp>12:
        print("Class Over")
    else: 
        print("Wait for class to finish!")
except KeyboardInterrupt:
    print("Cant Process")
except ValueError:
    print("Cant Process")
except AttributeError:
    print("Cant Process")
```

```python
while True:
    try:
        inp = int(input("Enter a number: "))
        if inp>12:
            print("Class Over")
        else: 
            print("Wait for class to finish!")
        break
    except ValueError:
        print("Cant Process")
```
