## Benchmarking

```python
# using https://docs.python.org/3/library/time.html
```

```python
import time

BATCH_SIZE = 1000000

start_time = time.time()
print(f"Start time: {start_time}")
for i in range(0, BATCH_SIZE):
    print(f"Evaluating (O){i}...")

end_time = time.time()
print(f"End time: {end_time}")
elapsed_time = end_time - start_time
elapsed_seconds = int(elapsed_time % 60)   # Remaining seconds
elapsed_minutes = int(elapsed_time // 60)  # Whole minutes
print(f"Total time: {elapsed_minutes}minutes {elapsed_seconds}seconds")
```
